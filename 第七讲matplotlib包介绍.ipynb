# Histogram 柱状图

```python
np.random.seed(2018)

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

n,bins,patches = plt.hist(x, 1000, normed=1, facecolor='b', alpha=0.5)#bins每个空间，patches表示柱状图，画x的值，画多少个柱状图，facecolor为蓝色，alpha透明度

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.text(60, .025, r'$mu=100,\  \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
n, bins, patches = plt.hist(weather_2012_final['Temp (C)'],100, facecolor='g', alpha=0.1)
print(n)
print(bins)
print(patches)
plt.grid(True)
plt.show()
```

## What is patches?

```python
x=np.random.normal(size=100)
print(x)

n,bins,patches=plt.hist(x)
plt.setp(patches[2], "facecolor" , 'r', alpha=0.5)
plt.show()
```

# Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi *(15*x)**2
plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
N = weather_2012_final.shape[0]
colors = np.random.rand(N)
plt.scatter(weather_2012_final["Temp (C)"], weather_2012_final['Dew Point Temp (C)'], c=colors, alpha=0.5)
plt.xlabel("Temp (C)")
plt.ylabel("Dew Point Temp(C)")
plt.title('Scatter of Temperture in tear 2012')
plt.show()
```

# Pie chart

```python
labels = 'Cats', 'Birds', 'Dogs', 'Pands'
sizes = [25,35,30,10]
explode = (0,0.8,0,0)#是否分开
fig1,ax1 = plt.subplots()
plt.pie(sizes, explode=explode, labels = labels, autopct='$1.1f%%', shadow=True, startangle=90)
#explde弹出距离，labels标题，autopct是百分值的位置，shadow阴影，startangle起始角度
ax1.axis('equal')#配比
plt.show()
```
